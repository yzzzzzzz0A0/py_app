* 字串 (Strings) 是由一連串的字元 (Characters) 組成的序列。
* Python 中的字串是**不可變的 (immutable)**。
* 您可以使用成對的單引號 (') 或雙引號 (") 來建立 Python 字串。

```python
# 多行字串可以使用連續三個單引號 (''') 或三個雙引號 (""") 來建立。
# 在字串內每一行結束的字元都會被保留。
# 如果你在一行的開頭或結尾使用空格，它們也都會被保留。
tool = '''This is a pan. 
This is a pencil. 
This is a book.
'''
print(tool)
# 你可以在使用雙引號的字串裡放入單引號，或是在使用單引號的字串裡放入雙引號
test = 'This is a "book".'
print(test)
```

# 使用 str() 進行型態轉換
您可以使用 `str()` 函式將其他資料型態轉換為字串。

```python
print(str(98.6))
print(str(1.0e4))
print(str(True))
print(str(1) + str(2))
```

# 使用 \ (反斜線) 進行轉義 (Escape)
* 在字元前加上反斜線 `\`，可以轉換字串內一些字元的涵義。
* 最常見的轉義序列是 `\n`，代表**換行**。
* 當字串本身被單引號包圍時，若內部需要顯示單引號，需使用 `\'`；雙引號同理使用 `\"`。
* 如果你需要字面上的反斜線，可以輸入兩個(用第一個來轉義第二個)
* **原始字串 (Raw string)** 會忽略這些跳脫符號的功能。

```python
print('12345678abc')
print('\tabc') # \t: Tab 符號 (縮排)

# 在字串中顯示引號及反斜線
print("This is \"special\".") 
print("This is a \\pen\\.")

# 原始字串 (r) 會忽略反斜線的預設功能
info = r'This is \n a row string' 
print(info)

# 原始字串不會取消真正的換行
test = r'''This is a pen. 
This is a pencil.'''
print(test)
```

# 使用 + 結合字串
* 您可以使用 `+` 運算子結合字串實體或字串變數。
* 您也可以直接將字串實體並排來結合它們（不適用於變數）。
* 當你串接字串時，Python並不會幫你加上空格。

```python
# 用 + 連結字串
test1 = "This is" + " the first test." 
print(test1)

# 直接並排字串
test2 = "This is" " the second test." 
print(test2)

 # 使用括號分行連接
test3 = ("This"
         " is" 
         " the"
         " third"
         " test.")
print(test3)

# 字串變數相加
a = "This is the"
b = " fourth test."
print(a + b) 

# print陳述式會幫每個引數間加入空格，並在結尾加入換行符號
a = 'This'
b = 'is'
c = 'the'
d = 'fifth'
e = 'test.'
print(a,b,c,d,e)
```

# 使用 [] 取得字元 (Indexing)
* 第一個（最左側）偏移量 (Offset) 為 0，下一個為 1，依此類推。
* 最後一個（最右側）偏移量可以用 -1 表示，向左則為 -2, -3，依此類推。

```python
name = "Jason"
print(name[-1])
```

# 字串是不可變的 (Immutable)
您無法直接插入字元或更改特定索引位置的字元。若要修改，通常需要建立新字串或使用 `replace()`或slice。

```python
name1 = "Henny"
# name1[0] = 'P' # 這會報錯，因為字串不可變
name2 = name1.replace("H", "P") 
print('name1 =', name1)
print('name2 =', name2)
```

```python
# 串列 (List) 結構是可變的
letter = ['a', 'b', 'c', 'd'] 
letter[1] = 'x'  
print(letter)
```

# 使用切片 (Slice) 取得子字串
您可以從字串中擷取一部分內容。切片格式為 `[開始:結束:間隔]`。
* `[:]` 擷取從頭到尾的整個序列。
* `[start:]` 從指定的開始位置直到結尾。
* `[:end]` 從開頭到**結束偏移量減 1**的位置。
* `[start:end]` 從開始位置到結束偏移量減 1 的位置。
* `[start:end:step]` 在範圍內按指定的間隔跳著擷取。
* 如果不指定 start，預設為 0；不指定 end，預設為字串長度。

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
print(letters[:])
print(letters[0:3:1])
print(letters[-3:])
print(letters[::-1]) # 快速反轉字串
print(letters[2:5])
```

```python
print(letters[18:-3])
print(letters[-6:-2])
```

```python
# 從開始到offset 20，間隔5
print(letters[:21:5])
```

```python
# 比字串開頭更前面的slice offset會被當成0
print(letters[-50:])
```

# 使用 len() 取得長度
回傳字串中的字元總數。

```python
len(letters)
```

# 使用 split() 拆分字串
根據指定的「分隔符號」將字串拆分成一個字串串列 (List)。

```python
days = 'Monday,Tuesday,Wednesday,Thursday,Friday,Saturday,Sunday'
day_list = days.split(',')
print(day_list)
print(day_list[1])
```

# 使用 join() 合併串列中的字串
與 split() 相反，`join()` 用一個字串作為接合劑，將串列中的元素黏起來：`'接合符號'.join(字串串列)`。

```python
separate_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
combined_days = '//'.join(separate_days)
print(combined_days)
```

# 使用 replace() 來替換
* 指定舊子字串、新子字串，以及要替代的次數。
* 注意：它會**回傳**修改後的新字串，但**不會修改**原本的字串變數內容。

```python
statement_1 = "This is a pan."
statement_2 = statement_1.replace('pan', 'pencil')
print(statement_1)
print(statement_2)
```

# 使用 strip() 剝除 (去頭去尾)
* 若不帶參數，預設會移除空白字元 (`' '`, `\t`, `\n`)。
* `strip()` 處理兩端，`lstrip()` 只處理左側，`rstrip()` 只處理右側。

```python
world = "   earth   "
print(world.strip())
print(world.lstrip())
print(world.rstrip())
```

```python
# 可以要求strip()移除多字元字串內的任何字元
blurt = "What the...!!??"
blurt.strip('.!?')
```

```python
# strip()只能移除頭尾的字元，如果要移除中間的字元，需要用replace()
world = "   earth   "
print(world.strip('r'))
```

# 搜尋與選擇 (Search and Select)
Python有大量的字串函式

```python
poem = '''All that doth flow we cannot liquid name
Or else would fire and water be the same;
But that is liquid which is moist and wet
Fire that property can never get.
Then 'tis not cold that doth the fire put out
But 'tis the wet that makes it die, no doubt.'''
print(poem.find("that"))   # 回傳第一個 "that" 的位置，找不到回傳 -1
print(poem.rfind("that"))  # 回傳最後一個 "that" 的位置，找不到回傳 -1
print(poem.index("that"))  # 回傳第一個 "that" 的位置，找不到則丟出例外 (Exception)
```

```python
# 這首詩有多少字元
print(len(poem))
# 它的開頭是不是All
print(poem.startswith('All'))
# 它的結尾是不是That's all, folks! ?
print(poem.endswith("That\'s all, folks!"))
# the 這個三字母序列出現幾次
print(poem.count("the"))
# 該字元是否只有字母及數字
print(poem.isalnum())
```

# 大小寫轉換 (Case)

```python
test = "this is a test!"
print(test.capitalize()) # 字首大寫
print(test.upper())      # 全部大寫
print(test.lower())      # 全部小寫
print(test.swapcase())   # 大小寫互換
```

# 字串格式化 (Formatting)
* 您可以將資料值插入字串中，用於產生報表或特定格式的輸出。
* Python 提供三種樣式：
  1. 舊式 (Old style, %)
  2. 新式 (New style, .format())
  3. f-string (Python 3.6+ 推薦)

# 舊式：%
* 轉換類型：  
    %s 字串 (string)  
    %d 十進制整數  
    %x 十六進制整數  
    %o 八進制整數  
    %f 十進位浮點數  
    %e 指數福點數  
    %g 十進制或整數浮點數   
    %% 顯示字元 %  

```python
name = "Jason"
print("My name is %s." % name)
print("My name is %12s." % name)   # 設定寬度為 12
print("My name is %-12s." % name)  # 靠左對齊
print("Your grade is %.2f." % 98.67433) # 四捨五入至小數點第二位
```

# 最新式：f-string
* 於 Python 3.6 加入，是目前最推薦的方式。
* 直接在開頭的引號前面輸入字母f或F。  
* 在大括號裡面加入變數名稱或運算式，來將他們的值放入字串。
* 對齊符號：`<` (左, 預設), `>` (右), `^` (置中)。
* 可指定最小寬度與精度。

```python
name = "Jason"
height = 175
print(f"My name is {name} and my height is {height}.")
print(f"My name is {name.upper():^20} and my height is {height:>20}.")
print(f"{name=}, {height=}") # 除錯時很好用：同時顯示變數名稱與數值
print(f"{name[2:]=}, {height/100=}") # 可以在其中使用運算式
```
