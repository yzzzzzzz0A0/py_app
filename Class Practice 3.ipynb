 # 隨機模組 (Random Module)
* 模組（Module）簡單來說就是一個 .py 檔案（例如：`my_module.py`）。
* 它可以包含函式、類別和變數。
* 在其他腳本或 Jupyter Notebook 中，使用 `import` 來匯入模組。
* 如果您在匯入後修改了 `my_module.py`，Jupyter 可能仍在使用舊版本。此時請使用 `importlib.reload` 重新載入：
  ```python
  import importlib  
  import my_module  
  importlib.reload(my_module)
  ```

````python
# 匯入自定義模組
import importlib  # 1. 先匯入工具包
import my_module  # 2. 匯入您的模組
importlib.reload(my_module) # 3. 每次執行這行時，都確保抓到最新版內容
print(my_module.my_number)
print(my_module.greet("Jason"))
````

````python
import random
random_int = random.randint(1, 10) # 顯示 1 到 10 之間的隨機整數
print(random_int)
random_float = random.uniform(1, 10) # 顯示 1 到 10 之間的隨機浮點數
print(random_float)
````

````python
# 隨機投擲硬幣
coin = random.randint(0, 1)
if coin == 1:
    print("正面 (head)")
else:
    print("反面 (tail)")
````

# 猜數字遊戲

````python
import random
com_number = random.randint(1,10)
count = 0
while True: 
    user_number = int(input("請輸入 1 到 10 之間的數字: "))
    if user_number > com_number:
        print('猜小一點')
        count += 1
    elif user_number < com_number:
        print('猜大一點')
        count += 1
    else:
        print('恭喜你猜對了！')
        break
    
    if count == 3:
        print('很遺憾，你已經猜了三次，遊戲結束！')
        print(f'正確的數字是 {com_number}')
        break
````

# 串列 (List)
Python 中的串列是一種**可變 (mutable)**、**有序**且**可索引**的集合，可以存儲多個項目，包括不同的資料型態。串列使用方括號 `[]` 定義。
* **有序 (Ordered)**：元素有固定的順序，可以透過索引存取。
* **可變 (Mutable)**：可以修改、新增或移除項目。
* **允許重複**：串列可以包含重複的元素。
* **異質性 (Heterogeneous)**：可以存儲不同的資料型態（例如：int, str, float 等）。

````python
# 第一個項目的索引為 0
# 可以使用負數索引
fruits = ["Cherry", "Apple", "Pear"]
print(fruits[-1])
fruits.append("Orange")  # 在串列末尾新增 "Orange"
print(fruits)
fruits.insert(1, "Banana") # 在 Cherry 之後新增 "Banana"
print(fruits)
print(len(fruits)) # 串列中的項目數量
````

````python
# 巢狀串列 (Nested List)
fruits = ["Cherry", "Apple", "Pear", "Orange"]
vegetable = ["Potato", "Spinach", "Carrot"]
box = [fruits, vegetable]
print(box[0][2]) # 存取 fruits 中的 "Pear"
````
